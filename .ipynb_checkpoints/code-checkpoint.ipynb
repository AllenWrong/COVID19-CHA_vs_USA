# 原生数据整理

* ## 刘培宇raw_data数据提取  (口罩)

```python
import pandas as pd
import re
dirname='data/Eng/口罩/'
start = 0
file = 'data/raw_data/口罩-English.csv'

df=pd.read_csv(file,encoding='utf-8',error_bad_lines=False)
```

```python
df.head(3)
```

```python
df.drop_duplicates('text',inplace=True)
```

```python
titles = df.title
times = df.time
texts = df.text

from tqdm import tqdm

for batch,(title,time,text) in enumerate(zip(titles,times,texts)):
    title=re.sub('[^a-zA-Z?\.,! ]','',title)
    text=re.sub('[^a-zA-Z?\.,! ]','',text)
    file=open(dirname+'%s.txt'%batch,'w')
    file.write(title.strip()+'\n')
    file.write(time+'\n')
    file.write(text)
    file.close()
    batch += 1
    if batch%20==0:
        print(batch)
```

* ## 关忠超raw_data数据提取

```python
import pandas as pd
import re
dirname='data/CHA/疫情/'
start = 0
file = 'data/raw_data/疫情-chinese.csv'

df=pd.read_csv(file)

```

```python
df.head(3)
```

```python
df.loc[:,'时间']=df.loc[:,'时间'].str.replace('\n',' ').str[:11].str.replace('  ','/')+df.loc[:,'时间'].str.replace('\n',' ').str[12:]
```

```python
df=df[df.loc[:,'时间'].str.len()<20]
```

```python
df.drop_duplicates('正文',inplace=True)
```

```python
titles = df.loc[:,'标题']
times = df.loc[:,'时间']
texts = df.loc[:,'正文']

from tqdm import tqdm

for batch,(title,time,text) in enumerate(zip(titles,times,texts)):
    
    try:
        title=re.sub('[^\u4e00-\u9fa5，。？！]','',title)
        title=re.sub('  ',' ',title)
        text=re.sub('[^\u4e00-\u9fa5，。？！]','',text)
    except:
        continue
    file=open(dirname+'%s.txt'%batch,'w')
    file.write(title.strip()+'\n')
    file.write(time+'\n')
    file.write(text)
    file.close()
    batch += 1
    if batch%20==0:
        print(batch)
```

* ## 石璞raw_data数据提取

```python
import pandas as pd
import re
dirname='data/Eng/疫苗/'
start = 0
file = 'data/raw_data/新冠疫苗_英文/china daily.csv'

df=pd.read_csv(file).loc[:,['标题','时间','正文']]


df.head()
```

```python

titles = df.loc[:,'标题']
times = df.loc[:,'时间']
texts = df.loc[:,'正文']

from tqdm import tqdm

for batch,(title,time,text) in enumerate(zip(titles,times,texts)):
    try:
        title=re.sub('[^a-zA-Z?\.,! ]','',title)
        text=re.sub('[^a-zA-Z?\.,! ]','',text).strip()
        time=re.sub('[^0-9- :]','',time).strip(':').strip()
    except:
        pass
    if type(text)!=str or len(text)<20:
        continue
    file=open(dirname+'%s.txt'%batch,'w')
    
    try:
        file.write(title.strip()+'\n')
        file.write(time+'\n')
        file.write(text)
        file.close()
    except:
        continue
    batch += 1
    if batch%20==0:
        print(batch)
```

* ## 无愧raw_data数据提取 (防疫)

```python
import pandas as pd
import re
dirname='data/Eng/防疫/'
start = 0
file = 'data/raw_data/防疫英文.csv'
df=pd.read_csv(file,encoding='gb18030')


df.head()
```

```python

titles = df.loc[:,'标题']
times = df.loc[:,'简介']
texts = df.loc[:,'正文']

from tqdm import tqdm

for batch,(title,time,text) in enumerate(zip(titles,times,texts)):
    try:
        time=re.sub('[^0-9- :]','',time).strip()
        title=re.sub('[^a-zA-Z?\.,! ]','',title)
        text=re.sub('[^a-zA-Z?\.,! ]','',text).strip()
    except:
        pass
    if type(text)!=str or len(text)<20:
        continue
    file=open(dirname+'%s.txt'%batch,'w')
    
    try:
        file.write(title.strip()+'\n')
        file.write(time+'\n')
        file.write(text)
        file.close()
    except:
        continue
    batch += 1
    if batch%20==0:
        print(batch)
```

# 所有数据加载

```python
import os
from jieba import posseg
from tqdm import tqdm
path = 'data/CHA/'



def get_words(filename):
    file=open(filename,'r')
    data=file.read().split('\n')
    text=data[-1]
    date=data[1]
    file.close()
    sentence_seged = posseg.cut(text)
    word=[]
    for x in sentence_seged:
        if x.flag=='n':
            word.append(x.word)
            
    return word,date



words =[]
date = []
for i in tqdm(os.listdir(path),ncols=90):
    if i[0]=='.':
        continue
    for file in os.listdir(path+i+'/')[:10]:
        if file[0]=='.':
            continue
        filename=path+i+'/'+file
        
        w,d=get_words(filename)
        if len(w)<20:
            continue
        words.append(w)
        date.append(d)

df=pd.DataFrame({'date':date,'content':words})
df.loc[:,'date']=pd.to_datetime(df.date)
df.set_index('date',inplace=True)
df.to_csv('data/useful_data.csv',encoding='gbk')
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_selection import chi2,SelectKBest
from sklearn.decomposition import LatentDirichletAllocation
import json
from matplotlib import pyplot as plt
from wordcloud import WordCloud
from sklearn.cluster import KMeans
plt.rcParams['font.sans-serif']=['Arial Unicode Ms']

    


def get_stop_words(filename):
    file=open(filename,'r',encoding='gbk')
    stop_words = file.read().split('\n')
    file.close()
    return stop_words


def get_plot(d,n_components,filename,cloud_num_words=100):
	"""
	d:              DataFrame ->  文件数据索引后的
	n_components:   int       ->  主题数
	filename:       str       ->  文件名,例如 '国内,2021-1～2021-7,'  
	cloud_num_words int       ->  每张词云显示词的个数

	return 
	"""
	stop_words = get_stop_words('data/stop_words.txt')
	data=d['content'].values.tolist()
	classifier = d['class'].tolist()
	cotVector = CountVectorizer(stop_words=stop_words)
	vector=cotVector.fit_transform(data)
	word_array = vector.toarray()
	feature_name = cotVector.get_feature_names()
	feature_name[:10]
	words = pd.Series(feature_name)
	lda = LatentDirichletAllocation(n_components=n_components,learning_offset=50,max_iter=50)
	docres = lda.fit_transform(word_array)

	model=KMeans(n_clusters=n_components)
	classifier=model.fit_predict(docres)
	f,p_value = chi2(word_array,classifier)
	model=SelectKBest(chi2,k=400)
	new_word_array = model.fit_transform(word_array,classifier)
	words=words[model.get_support()]
	words=pd.Series(words.tolist())
	lda = LatentDirichletAllocation(n_components=n_components,learning_offset=50,max_iter=50)
	docres = lda.fit_transform(new_word_array)

	for t_index in range(n_components):
		index = lda.components_[t_index].argsort()[-60:]
		ws = words[index].tolist()
		i = lda.components_[t_index][index]
		cloud = WordCloud(font_path='/system/library/fonts/Hiragino Sans GB.ttc')
		s=cloud.fit_words(dict(zip(ws,i)))

		plt.figure(figsize=(12,12))
		cloud.to_file(filename+'t_index.png')


def main(method,n_components,cloud_num_words=100):
	method='年'
	df=pd.read_csv('data/useful_data.csv',encoding='gbk')
	df.loc[:,'date']=pd.to_datetime(df.date)
	df.set_index('date',inplace=True)
	slic = []
	if method=='季':
	    slic.append(['2020-1','2020-4'])
        slic.append(['2020-4','2020-7'])
        slic.append(['2020-7','2020-10'])
        slic.append(['2020-10','2021-1'])
        slic.append(['2021-1','2021-4'])
        slic.append(['2021-4','2021-7'])
        slic.append(['2021-7','2021-12'])
    elif method=='半年':
        slic.append(['2020-1','2020-7'])
        slic.append(['2020-7','2021-1'])
        slic.append(['2021-1','2021-7'])
        slic.append(['2021-7','2021-12'])
    elif method=='年':
        slic.append(['2020-1','2021-1'])
        slic.append(['2021-1','2021-12'])
    for i in slic:
        print(i)
        d=df.loc[i[0]:i[1],:]
        date = i[0]+'~'+i[1]
        get_plot(d,n_components,'Img/国内,%s,'%date)

if __name__=="__main__":
	main(3)
```
